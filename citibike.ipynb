```python
import pandas as pd
```

```python
# Load the CSV file
df = pd.read_csv('data/202003-citibike-tripdata_1.csv')
```

```python
df.dtypes
```

```python
df['started_at'].sort_values()
```

```python
filtered_df=df.query("started_at>='2020-03-01' and started_at<='2020-04-01'")
filtered_df['started_at'].sort_values()
```

```python
filtered_df
```

```python
# Check for missing values
print(filtered_df.isnull().sum())
```

```python
# Drop rows with missing data
df_cleaned = filtered_df.dropna()
```

```python
df_cleaned['ended_at'] = df_cleaned['ended_at'].astype('datetime64[ms]')
df_cleaned['started_at'] = df_cleaned['started_at'].astype('datetime64[ms]')
```

```python
df_cleaned.dtypes
```

```python
df_cleaned['trip_duration'] = df_cleaned['ended_at'] - df_cleaned['started_at']
df_cleaned
```

```python
df_cleaned['trip_duration'].sort_values()
```

```python
df_cleaned.to_csv('202003-citibike-tripdata_cleaned.csv', index=False)
```
